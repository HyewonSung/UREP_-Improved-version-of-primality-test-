```python
# n의 prime factor의 리스트를 구하는 함수
def prime_factorization(n):
    factors=[]
    count=1
    i=2
    
    while i<=n/i :
        while n%i==0 :
            factors.append(i)
            n=n/i
            count=count+1
        i=i+1
    
    if n>1:
        factors.append(n)
        count=count+1
        
    factors.append(count-1) #prime factors의 개수
    return factors    
```

```python
# n이 (2x+1)(4x+1) 꼴인지 판별하는 함수
def isTheFormOf_2xp1_4xp1(n):
    prime_factors = prime_factorization(n)
    
    #prime factor의 개수가 2가 아닌 경우
    if (prime_factors[-1]!=2):
        #print(n,"is not the form of (2x+1)(4X+1), because the number of prime factors of",n,"is not 2.")
        return False
    
    #p,q가 2x+1,4x+1 form인지 확인하기
    p=prime_factors[0]; q=prime_factors[1]
    if ((p - 1) % 2 == 0): #p가 2x+1 form인 경우
        x = (p - 1) / 2
        if ((4 * x + 1) == q):  #q가 4x+1 form인 경우
            #print(n,"is the form of (2x+1)(4X+1)")
            return True
        
        else : #q가 4x+1 form이 아닌 경우
            #print(n,"is not the form of (2x+1)(4X+1), because q is not the form of 4x+1.")
            return False
        
    
    else : #p가 2x+1 form이 아닌 경우
        #print(n,"is not the form of (2x+1)(4X+1), because p is not the form of 2x+1.")
        return False
```

```python
# n이 (2x+1)(6x+1) 꼴인지 판별하는 함수
def isTheFormOf_2xp1_6xp1(n):
    prime_factors = prime_factorization(n)
    
    #prime factor의 개수가 2가 아닌 경우
    if (prime_factors[-1]!=2):
        #print(n,"is not the form of (2x+1)(6X+1), because the number of prime factors of",n,"is not 2.")
        return False
    
    #p,q가 2x+1,6x+1 form인지 확인하기 
    p=prime_factors[0]; q=prime_factors[1]
    if ((p - 1) % 2 == 0): #p가 2x+1 form인 경우
        x = (p - 1) / 2
        if ((6 * x + 1) == q):  #q가 6x+1 form인 경우
            #print(n,"is the form of (2x+1)(6X+1)")
            return True
        
        else : #q가 6x+1 form이 아닌 경우
            #print(n,"is not the form of (2x+1)(6X+1), because q is not the form of 6x+1.")
            return False
        
    
    else : #p가 2x+1 form이 아닌 경우
        #print(n,"is not the form of (2x+1)(6X+1), because p is not the form of 2x+1.")
        return False
```

```python
def powmod(x,e,m): # x^e mod m 구하는 함수
    X = x
    E = e
    Y = 1
    while E > 0:
        if E % 2 == 0:
            X = (X * X) % m
            E = E/2
        else:
            Y = (X * Y) % m
            E = E - 1
    return Y

def miillerTest(d, n): 
    a = 2 + random.randint(1, n - 4) # a 는 [2..n-2] 안의 랜덤 정수  
    x = powmod(a, d ,n) # x = a^d % n 
    
    if (x == 1 or x == n - 1): 
        return True
  
    while (d != n - 1): 
        x = (x * x) % n
        d *= 2
        if (x == 1): 
            return False
        if (x == n - 1): 
            return True
    return False

def Original_Miller_Rabin(n, reps):
     # Corner cases 
    if (n <= 1 or n == 4): 
        return False
    if (n <= 3): 
        return True
    
    d = n - 1
    while (d % 2 == 0): 
        d //= 2
  
    for i in range(reps): # reps만큼 Miller Rabin test 돌리기
        if (miillerTest(d, n) == False): 
            return False
  
    return True


def Refined_Miller_Rabin(n, reps):
    
    ## Remove 3 tyoes of the forms
    if isTheFormOf_2xp1_4xp1(n):  return False
    if isTheFormOf_2xp1_6xp1(n):  return False
    if n in pqr_list:  return False
    
    ##여기부터는 원래의 Miller Rabin test와 같음
    Original_Miller_Rabin(n, reps)
```

```python
import random

random_numbers=[]  # test set 만들기
for i in range(1000):
    number=random.randint(1000000000000,100000000000000000) #10^10 ~ 10^15
    random_numbers.append(number)
```

```python
#정확성 비교 1 - test set이 랜덤정수리스트

##Original Miller Rabin
true=0;false=0
for n in random_numbers:
    if Original_Miller_Rabin(n, 1):
        true=true+1
    else: false=false+1
print("Original : true(probably prime)=",true,"/ false(composite)=",false)

##Refined Miller Rabin
true=0;false=0
for n in random_numbers:
    if Refined_Miller_Rabin(n, 1):
        true=true+1
    else: false=false+1
print("Refined : true(probably prime)=",true,"/ false(composite)=",false)
```
